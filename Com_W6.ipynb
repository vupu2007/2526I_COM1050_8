```python
#W6A9
numbers = list(map(int, input("Nhập danh sách các số nguyên, cách nhau bằng dấu cách: ").split()))
k = int(input("Nhập giá trị k : "))
pairs = []
for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] + numbers[j] == k:
            pairs.append((numbers[i], numbers[j]))
pairs.sort()
print("Các cặp số có tổng bằng", k, "là:")
for pair in pairs:
    print(pair)
```

```python
#W6A10
dict1 = eval(input("Nhập dictionary thứ nhất: "))
dict2 = eval(input("Nhập dictionary thứ hai: "))
for key, value in dict2.items():
    if key in dict1:
        dict1[key] += value
    else:
        dict1[key] = value
dict1 = dict(sorted(dict1.items()))
print(dict1)
```

```python
#W6A11
nums = tuple(map(int, input().split()))

even_nums = ()
odd_nums = ()

for x in nums:
    if x % 2 == 0:
        even_nums += (x,)
    else:
        odd_nums += (x,)

print(even_nums)
print(odd_nums)
```

```python
#W6A12
nums = list(map(int, input().split()))
max_count = 0
result = nums[0]
for x in nums:
    count = nums.count(x)
    if count > max_count or (count == max_count and x < result):
        max_count = count
        result = x
print(result)
```

```python
#W6A14
n = int(input())
a = [list(map(int, input().split())) for _ in range(n)]
m = int(input())
b = [list(map(int, input().split())) for _ in range(m)]
set_a = set(sum(a, []))
set_b = set(sum(b, []))
result = list(set_a & set_b)
print(sorted(result))
```

```python
#W6A15
n = int(input())
d = {}
while True:
    line = input().strip()
    if not line:
        break
    k, v = line.split()
    d[k] = int(v)
new_d = {k: v for k, v in d.items() if v > n}
print(new_d)
```
