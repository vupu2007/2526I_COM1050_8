```python
#W5A1
def max2(a, b):
    if a > b:
        return a
    elif b > a:
        return b
    else:
        return "Hai số bằng nhau"
m = input("Nhập số a,b: ").split(",")
a = int(m[0])
b = int(m[1])
print("Số lớn hơn là:", max2(a, b))
```

```python
#W5A2
def swap(a, b):
    return b, a
so = input("Nhập hai số a và b: ").split()
a = int(so[0])
b = int(so[1])
print(swap(a, b))
```

```python
#W5A3
def snt(n):
 if n < 2:
    return False
 elif n == 2 :
    return True
 else:
    for i in range(2, n):
        if n % i == 0:
             return False
        else :
           return True
a = int(input("Nhạp số nguyên a : "))
if snt(a):
    print("Là số nguyên tố")
else :
    print("Không phải là số nguyên tố")
```

```python
#W5A4
def shh(a):
  tong = 0
  for i in range(1,a//2 + 1):
    if a % i == 0 :
      tong += i
    if tong == a :
      return True
  return False
a = int(input("Nhập số nguyên a : "))
if shh(a):
  print(a,"là số hoàn hảo ")
else :
  print(a,"không phải là số hoàn hảo ")
```

```python
#W5A5
def find_first_index(nums, k):
    for i, value in enumerate(nums):
        if value == k:
            return i
    return -1
nums = list(map(int, input("Nhập list số nguyên (cách nhau bằng dấu cách): ").split()))
k = int(input("Nhập số nguyên dương k: "))
result = find_first_index(nums, k)
print("Vị trí xuất hiện đầu tiên của", k, "là:", result)
```

```python
#W5A6
def giaithua(n):
  tich = 1
  for i in range(1,n+1):
    tich = tich * i
  return tich
n= int(input("Nhập số nguyên dương n : "))
print("Giai thừa của n là : ", giaithua(n))
```

```python
#W5A7
num1 = float(input("Nhập số thứ nhất : "))
op   = input("Nhập phép tính : ")
num2 = float(input("Nhập số thứ hai : "))

if op == '+':
    kq = num1 + num2
elif op == '-':
    kq = num1 - num2
elif op == '*':
    kq = num1 * num2
elif op == '/':
    kq = num1 / num2

print(round(kq, 2))
```

```python
#W5A8
def hamming_distance(x, y):
    count = 0
    while x > 0 or y > 0:
        bit_x = x % 2
        bit_y = y % 2
        if bit_x != bit_y:
            count += 1
        x = x // 2
        y = y // 2
    return count
x = int(input("Nhập số thứ nhất: "))
y = int(input("Nhập số thứ hai: "))
kq = hamming_distance(x, y)
print("Khoảng cách Hamming là:", kq)
```

```python
#W5A9
def tong_chu_so(n):
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10
    return tong
n = int(input("Nhập n: "))
print(tong_chu_so(n))
```

```python
#W5A10
def isomorphic(a, b):
    if len(a) != len(b):
        return False

    map_a = {}
    map_b = {}

    for x, y in zip(a, b):
        if x in map_a and map_a[x] != y:
            return False
        if y in map_b and map_b[y] != x:
            return False
        map_a[x] = y
        map_b[y] = x

    return True

# Phần chính: nhập dữ liệu
a = input("Nhập từ thứ nhất: ")
b = input("Nhập từ thứ hai: ")

# Kiểm tra và in kết quả
if isomorphic(a, b):
    print("Hai từ đẳng cấu (True)")
else:
    print("Hai từ không đẳng cấu (False)")
```
