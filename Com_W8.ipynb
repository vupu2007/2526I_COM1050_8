```python
  #W8A1
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h

    def tinh_chu_vi(self):
        return (2 * 3.14 * self.r * self.h)

    def tinh_the_tich(self):
        return (3.14 * self.r * self.r * self.h)

# Nhập dữ liệu
a = int(input("Nhập bán kính r: "))
b = int(input("Nhập chiều cao h: "))
hinh_tru = Cylinder(a, b)
print("Diện tích xung quanh:", format(hinh_tru.tinh_chu_vi(),'.2f'))
print("Thể tích:", format(hinh_tru.tinh_the_tich(),'.2f') )
```

```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)

    def is_valid(self):
        if self.year <= 0 or self.month < 1 or self.month > 12 or self.day < 1:
            return False

        days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

        if self.month == 2 and self.is_leap_year():
            return self.day <= 29

        return self.day <= days_in_month[self.month]

    def next_day(self):
        days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

        if self.month == 2 and self.is_leap_year():
            days_in_month[2] = 29

        self.day += 1
        if self.day > days_in_month[self.month]:
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

    def __str__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"

s = input().strip()

try:
    d, m, y = map(int, s.split('/'))
    date = Date(d, m, y)

    if not date.is_valid():
        print("INVALID")
    else:
        date.next_day()
        print(date)
except:
   print("INVALID")
```

```python
#W8A3
class toado:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def goctoado(self):
        if self.x == 0 and self.y == 0:
            return "M là gốc tọa độ"
        return None

    def tructung(self):
        if self.x == 0 and self.y != 0:
            return "M thuộc trục tung"
        return None

    def truchoanh(self):
        if self.y == 0 and self.x != 0:
            return "M thuộc trục hoành"
        return None

    def khoangcach(self):
        d = (self.x**2 + self.y**2)**(1/2)
        return f"Khoảng cách đến gốc tọa độ: {d:.2f}"

M_input = input("Nhập tọa độ M (x,y): ")
try:
    x_val, y_val = map(int, M_input.split(","))
    diem_M = toado(x_val, y_val)


    ket_qua = diem_M.goctoado() or diem_M.tructung() or diem_M.truchoanh()
    if ket_qua:
        print(ket_qua)
    else:
        print("M nằm trong các hệ phần tư")

    print(diem_M.khoangcach())

except :
    print("Không hợp lệ. Vui lòng nhập đúng định dạng x,y (Ví dụ: 3,4)")


```

```python
#W8A4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def set_numbers(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def product(self):
        return self.a * self.b

    def divide(self):
        if self.b == 0:
            raise ZeroDivisionError("Không thể chia cho số 0!")
        return self.a / self.b

    def power(self):
        return self.a ** self.b

    def mod(self):
        if self.b == 0:
            raise ZeroDivisionError("Không thể chia lấy dư cho số 0!")
        return self.a % self.b

def show_menu():
    print("\n--- MENU MÁY TÍNH ---")
    print("1. Cộng (+)")
    print("2. Trừ (-)")
    print("3. Nhân (*)")
    print("4. Chia (/)")
    print("5. Lũy thừa (^)")
    print("6. Chia lấy dư (%)")
    print("7. Cập nhật 2 số mới")

try:
    num1 = float(input("Nhập số thứ nhất: "))
    num2 = float(input("Nhập số thứ hai: "))
    calc = Calculator(num1, num2)
except Exception as e:
    print(f"Lỗi nhập liệu ban đầu: {e}")
    exit()

while True:
    show_menu()
    choice = input("Chọn phép tính (1-7): ")

    try:
        if choice == '1':
            print(f"Kết quả: {calc.add()}")
        elif choice == '2':
            print(f"Kết quả: {calc.subtract()}")
        elif choice == '3':
            print(f"Kết quả: {calc.product()}")
        elif choice == '4':
            print(f"Kết quả: {calc.divide()}")
        elif choice == '5':
            print(f"Kết quả: {calc.power()}")
        elif choice == '6':
            print(f"Kết quả: {calc.mod()}")
        elif choice == '7':
            n1 = float(input("Nhập số thứ nhất mới: "))
            n2 = float(input("Nhập số thứ hai mới: "))
            calc.set_numbers(n1, n2)
            print("Đã cập nhật số mới thành công!")
        else:
            print("Lựa chọn không hợp lệ!")

    except Exception as e:
        print(f"Có lỗi xảy ra: {e}")

    quit_answer = input("Bạn có muốn thoát? (Gõ 'yes' để thoát): ").lower()
    if quit_answer == "yes":
        print("Cảm ơn bạn đã sử dụng máy tính!")
        break
```

```python
#W8A5
import sys

class ShoppingCart:
    def __init__(self):
        self.ds = []

    def add_item(self, name, price):
        item = {"name": name, "price": price}
        self.ds.append(item)
        print(f"Đã thêm '{name}' vào giỏ hàng.")

    def remove_item(self, name):
        for item in self.ds:
            if item["name"] == name:
                self.ds.remove(item)
                print(f"Đã xóa '{name}' khỏi giỏ hàng.")
                return
        print(f"Không tìm thấy sản phẩm '{name}' để xóa.")

    def check_item(self):
        if not self.ds:
            return "Giỏ hàng trống"
        return f"Giỏ hàng đang có {len(self.ds)} sản phẩm"

    def total(self):
        tong = sum(item["price"] for item in self.ds)
        return tong

    def show_items(self):
        if not self.ds:
            print("Giỏ hàng chưa có gì.")
        else:
            for i, item in enumerate(self.ds, 1):
                print(f"{i}. {item['name']} - Giá: {item['price']}")

    def xoa_tat_ca(self):
        self.ds.clear()
        print("Đã xóa toàn bộ giỏ hàng.")

# --- CHƯƠNG TRÌNH CHÍNH ---
cart = ShoppingCart()

while True:
    print("\n--- MENU QUẢN LÝ GIỎ HÀNG ---")
    print("1. Thêm sản phẩm")
    print("2. Xóa sản phẩm theo tên")
    print("3. Kiểm tra giỏ hàng")
    print("4. Tính tổng tiền")
    print("5. Hiển thị danh sách")
    print("6. Xóa sạch giỏ hàng")

    try:
        choice = input("Chọn một thao tác (1-6): ")

        if choice == '1':
            name = input("Nhập tên sản phẩm: ")
            price = float(input("Nhập giá sản phẩm: "))
            cart.add_item(name, price)
        elif choice == '2':
            name = input("Nhập tên sản phẩm muốn xóa: ")
            cart.remove_item(name)
        elif choice == '3':
            print(cart.check_item())
        elif choice == '4':
            print(f"Tổng tiền: {cart.total()}")
        elif choice == '5':
            cart.show_items()
        elif choice == '6':
            cart.xoa_tat_ca()
        else:
            print("Lựa chọn không hợp lệ.")

        # Câu hỏi thoát chương trình theo đề bài
        confirm = input("Có muốn thoát khỏi chương trình không? (nhập 'yes' để thoát): ").lower()
        if confirm == "yes":
            print("Cảm ơn bạn đã sử dụng dịch vụ!")
            sys.exit(0) # Thoát an toàn

    except Exception as e:
        print(f"Lỗi: {e}")


```

```python
#W8A7
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không thể bằng 0")

        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common

        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"

def main():
    try:
        data = input("Nhập dòng dữ liệu (ví dụ 1/2 + 1/3 nhập là 1 2 + 1 3) ").split()
        if len(data) < 5:
            return None

        a, b, op, c, d = int(data[0]), int(data[1]), data[2], int(data[3]), int(data[4])

        f1 = Fraction(a, b)
        f2 = Fraction(c, d)

        if op == '+':
            result = f1.add(f2)
        elif op == '-':
            result = f1.sub(f2)
        elif op == '*':
            result = f1.mul(f2)
        elif op == '/':
            result = f1.div(f2)

        print(result)
    except Exception as e:
        pass

if __name__ == "__main__":
    main()
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def __str__(self):
        return f"Chủ thẻ: {self.owner}, Số dư: {self.balance}"

    def deposit(self, amount):
        print(f"Đã gửi: {amount}")
        return BankAccount(self.owner, self.balance + amount)

    def withdraw(self, amount):
        if self.balance < amount:
            print("Lỗi: Số dư không đủ!")
            return self
        print(f"Đã rút: {amount}")
        return BankAccount(self.owner, self.balance - amount)

s = input("Nhập tên và số dư ban đầu (ví dụ: An,1000): ").strip()
name, bal = s.split(',')
bank = BankAccount(name.strip(), int(bal))

n = int(input("Nhập số bước: "))
while n > 0:
    choice = input("1-Gửi, 2-Rút: ").strip()
    if choice == "1":
        amount = int(input("Số tiền gửi: "))
        bank = bank.deposit(amount)
    elif choice == "2":
        amount = int(input("Số tiền rút: "))
        bank = bank.withdraw(amount)

    print("Thông tin hiện tại:", bank)
    n -= 1

exit()



```

```python
#W8A9
class student:
    def __init__(self):
        self.ds = []

    def add_score(self, subject, score):
        monhoc = {"Subject": subject, "Score": score}
        self.ds.append(monhoc)
        return f"{subject} có số điểm: {score}"

name = input("Nhập tên học sinh: ")
studen = student()
n_original = int(input("Nhập số môn học: "))
n = n_original

while n > 0:
    sub = input("Nhập tên môn học: ")
    score = float(input("Nhập điểm môn học: "))
    studen.add_score(sub, score)
    n -= 1

if n_original > 0:
    total_score = sum(monhoc["Score"] for monhoc in studen.ds)
    average = total_score / n_original
    if average >= 8 :
      jud = "Exellent"
    elif 6.5 <= average < 8 :
      jud = "Good"
    elif 5 <= average < 6.5 :
      jud = " Average"
    else :
      jud = "Poor"
    print(f"{name}, trung bình môn: {average:.2f} - {jud}")

else:
    print("Không có môn học nào được nhập.")
```

```python
#w8a10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        count = sum(1 for book in self.books if book.author == name)
        return count

    def find_by_year(self, y):
        count = sum(1 for book in self.books if int(book.year) == int(y))
        return count

lib = Library()

try:
    line1 = input().split()
    if not line1:
        exit()
    n = int(line1[0])

    for _ in range(n):
        command_line = input().split(maxsplit=1)
        if not command_line:
            continue

        cmd = command_line[0]

        if cmd == "ADD":
            details = command_line[1].split(';')
            title = details[0].strip()
            author = details[1].strip()
            year = details[2].strip()

            new_book = Book(title, author, year)
            lib.add(new_book)

        elif cmd == "COUNT":
            author_name = command_line[1].strip()
            print(lib.count_by_author(author_name))

        elif cmd == "COUNTYEAR":
            year_query = command_line[1].strip()
            print(lib.find_by_year(year_query))

except EOFError:
    pass
```

```python
#W8A11
import math

class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def cross(self, v):
        return Vector(
            self.y * v.z - self.z * v.y,
            self.z * v.x - self.x * v.z,
            self.x * v.y - self.y * v.x
        )
    def dot(self, v):
        return self.x * v.x + self.y * v.y + self.z * v.z

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

def solve():
    pts = []
    for i in range(4):
        try:
            point = list(map(float, input().split()))
            pts.append(point)
        except EOFError:
            break

    if len(pts) < 4:
        return

    A, B, C, D = pts[0], pts[1], pts[2], pts[3]

    AB = Vector(B[0] - A[0], B[1] - A[1], B[2] - A[2])
    BC = Vector(C[0] - B[0], C[1] - B[1], C[2] - B[2])
    CD = Vector(D[0] - C[0], D[1] - C[1], D[2] - C[2])

    n1 = AB.cross(BC)
    n2 = BC.cross(CD)

    mau_so = n1.magnitude() * n2.magnitude()

    if mau_so == 0:
        print("0.00")
        return

    cos_phi = n1.dot(n2) / mau_so

    cos_phi = max(-1.0, min(1.0, cos_phi))

    phi = math.degrees(math.acos(cos_phi))

    print(f"{phi:.2f}")

if __name__ == "__main__":
    solve()
```

```python
#W8A12
class ComplexNumber:
    def __init__(self, initRealPart, initImaginaryPart):
        self.realPart = initRealPart
        self.imaginaryPart = initImaginaryPart

    def multiply(self, other):
        resReal = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        resImaginary = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(resReal, resImaginary)

print("--- Chương trình nhân hai số phức ---")

input1 = input("Nhập phần thực và ảo của số thứ nhất (cách nhau bởi dấu cách): ")
parts1 = input1.split()
num1 = ComplexNumber(float(parts1[0]), float(parts1[1]))

input2 = input("Nhập phần thực và ảo của số thứ hai (cách nhau bởi dấu cách): ")
parts2 = input2.split()
num2 = ComplexNumber(float(parts2[0]), float(parts2[1]))

result = num1.multiply(num2)

print("\nKết quả tích hai số phức là:")
print(f"Phần thực: {result.realPart}")
print(f"Phần ảo: {result.imaginaryPart}")
print(f"Dạng số phức: {result.realPart} + {result.imaginaryPart}i")
```

```python
#W8A13
import math

class Triangle:
    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.edge1 = initEdge1
        self.edge2 = initEdge2
        self.edge3 = initEdge3

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3
        if a + b > c and a + c > b and b + c > a and a > 0 and b > 0 and c > 0:
            p = (a + b + c) / 2
            area = math.sqrt(p * (p - a) * (p - b) * (p - c))
            return area
        else:
            return -1


try:
    userInput = input("Nhập độ dài 3 cạnh của tam giác (cách nhau bởi dấu cách): ")
    edges = list(map(int, userInput.split()))

    if len(edges) == 3:
        tri = Triangle(edges[0], edges[1], edges[2])
        area = tri.getArea()

        if area == -1:
            print("invalid")
        else:
            print(f"{area:.2f}")
    else:
        print("invalid")
except:
    print("invalid")
```
