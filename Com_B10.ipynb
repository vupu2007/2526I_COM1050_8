```python
#W10A2
def n(ds):
    i = 0
    while i < len(ds):
        if ds[i].isdigit():
            i += 1
        else:
            return None
    return "Danh sách hợp lệ"

ds = [x.strip() for x in input("Nhập danh sách: ").split(',')]

try:
    result = n(ds)
    print(result)
except Exception as e:
    print(f"Lỗi hệ thống: {e}")
```

```python
#W10A3
def is_happy(n):
    seen = set()
    while n != 1:
        if n in seen:
            return False
        seen.add(n)
        n = sum(int(digit) ** 2 for digit in str(n))
    return True


try:
    n = int(input("Nhập số nguyên dương: "))
    if n <= 0:
        raise ValueError

    if is_happy(n):
        print(n, "là số hạnh phúc")
    else:
        print(n, "không phải là số hạnh phúc")

except ValueError:
    print("Vui lòng nhập một số nguyên dương hợp lệ.")
```

```python
#W10A4
def tinh_tong():
    try:
        num1 = float(input("Nhập số thứ nhất: "))
        num2 = float(input("Nhập số thứ hai: "))

    except ValueError:
        print("Lỗi: Bạn đã nhập sai kiểu dữ liệu! Vui lòng nhập số.")

    else:
        tong = num1 + num2
        print(f"Tổng của {num1} và {num2} là: {tong}")

    finally:
        print("Kết thúc chương trình.")
tinh_tong()
```

```python
#W10A5
n = input("Nhập các phần tử (cách nhau bằng khoảng trống): ").split()
s = []

for item in n:
    try:
        so_nguyen = int(item)
        s.append(so_nguyen)
    except ValueError:
        print(f"Cảnh báo: '{item}' không thể chuyển sang int và sẽ bị bỏ qua.")

print("Danh sách các số nguyên sau khi chuyển:", s)
```

```python
#W10A10
try:
    n = int(input("Nhập số lượng phần tử N: "))
    if n <= 0:
        raise ValueError("N phải lớn hơn 0")

    print(f"Nhập {n} số nguyên dương (cách nhau bởi khoảng trắng):")
    mang = list(map(int, input().split()))

    if len(mang) != n:
        raise ValueError(f"Bạn phải nhập chính xác {n} phần tử.")

except ValueError as e:
    print(f"Lỗi nhập liệu: {e}")

else:
    if len(set(mang)) == len(mang):
        tong = sum(mang)
        print(f"Tổng các phần tử là: {tong}")
    else:
        print("Mang khong hop le")

finally:
    print("--- Chương trình kết thúc ---")
```

```python
#W10A11
def check_regex_manual():
    try:
        line = input().strip()
        if not line: return
        t = int(line)

        for _ in range(t):
            s = input().strip()
            is_valid = True
            if s and s[0] in "*+?{":
                is_valid = False

            repeat_chars = "*+?"
            for i in range(len(s) - 1):
                if s[i] in repeat_chars and s[i+1] in repeat_chars:
                    is_valid = False
                    break

            if s.count('(') != s.count(')') or s.count('[') != s.count(']'):
                is_valid = False

            if is_valid:
                print("True")
            else:
                print("False")

    except EOFError:
        pass
    except ValueError:
        pass

check_regex_manual()
```
