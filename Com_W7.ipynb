```python
#W7A1
def binary_search(arr, value):
    left, right = 0, len(arr) - 1
    while left <= right:  # while there is a valid range to check
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid  # Found the value
        elif arr[mid] < value:
            left = mid + 1  # search in right half
        else:
            right = mid - 1  # search in left half

    return -1  # Value not found
```

```python
#W7A2
def count_occurrences(arr,x):
  count = 0
  for i in arr:
    if i == x:
      count += 1
  return count
arr = input().split()
x = input()
print(f"Số lần xuất hiện của {x} là :",count_occurrences(arr,x))
```

```python
#W7A3a
def bubble_sort(arr):
  for i in range(len(arr)):
    swapped = False
    for j in range(0,len(arr) - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
        swapped = True
    if not swapped:
      break
  return arr
arr = input().split()
print(bubble_sort(arr))
```

```python
#W7A3b
def selection_sort(arr):
  for i in range(len(arr)):
    min_idx = i
    for j in range(i + 1, len(arr)):
      if arr[j] < arr[min_idx]:
        min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]
  return arr
arr = input().split()
print(selection_sort(arr))
```

```python
#W7A4
def tim_phan_tu_pho_bien_nhat(danh_sach):
    if not danh_sach:
        return None
    dem = {}
    for x in danh_sach:
        if x in dem:
            dem[x] += 1
        else:
            dem[x] = 1

    ket_qua = danh_sach[0]
    max_lan_xuat_hien = 0

    for x in danh_sach:
        if dem[x] > max_lan_xuat_hien:
            max_lan_xuat_hien = dem[x]
            ket_qua = x
    return ket_qua
ds = [1, 3, 2, 3, 2, 1, 4, 2, 3]
print("Phần tử xuất hiện nhiều nhất và sớm nhất là:", tim_phan_tu_pho_bien_nhat(ds))
```
