```python
#W7A1
def binary_search(arr, value):
    left, right = 0, len(arr) - 1
    while left <= right:  # while there is a valid range to check
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid  # Found the value
        elif arr[mid] < value:
            left = mid + 1  # search in right half
        else:
            right = mid - 1  # search in left half

    return -1  # Value not found
```

```python
#W7A3a
def bubble_sort(arr):
  for i in range(len(arr)):
    swapped = False
    for j in range(0,len(arr) - i - 1):
      if arr[j] > arr[j + 1]:
        arr[j], arr[j + 1] = arr[j + 1], arr[j]
        swapped = True
    if not swapped:
      break
  return arr
arr = input().split()
print(bubble_sort(arr))
```

```python
#W7A3b
def selection_sort(arr):
  for i in range(len(arr)):
    min_idx = i
    for j in range(i + 1, len(arr)):
      if arr[j] < arr[min_idx]:
        min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]
  return arr
arr = input().split()
print(selection_sort(arr))
```

```python
#W7A5
def dem_cap(ar, x):
    dem = 0
    n = len(ar)
    for i in range(n):
        for j in range(i + 1, n):
            if ar[i] + ar[j] == x:
                dem += 1
                print(f"Cặp tìm thấy: ({ar[i]}, {ar[j]})")

    return dem

mang = [1, 2, 3, 4, 5]
tong_x = 5
ket_qua = dem_cap(mang, tong_x)
print(f"==> Tổng cộng có {ket_qua} cặp.")
```

```python
#W7A7
def tim_day_con_dai_nhat(nums):
    if not nums: return 0
    L = [1] * len(nums)

    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                if L[j] + 1 > L[i]:
                    L[i] = L[j] + 1
    return max(L)
nums = [ 5, 2, 3, 4]
print("Độ dài lớn nhất:", tim_day_con_dai_nhat(nums))
```

```python
#W7A8
def giai_bai_toan(intervals, queries):
    min_len_map = {}
    intervals.sort(key=lambda x: (x[1] - x[0] + 1), reverse=True)
    for l, r in intervals:
        do_dai = r - l + 1
        for thoi_gian in range(l, r + 1):
            min_len_map[thoi_gian] = do_dai
    results = []
    for q in queries:
        results.append(min_len_map.get(q, -1))

    return results
intervals = [[1, 4], [2, 3]]
queries = [2, 5]
print(giai_bai_toan(intervals, queries))
```

```python
#W7A9
def count_smaller_binary_search(nums):
    result = []
    sorted_list = []
    for i in range(len(nums) - 1, -1, -1):
        x = nums[i]
        left = 0
        right = len(sorted_list)

        while left < right:
            mid = (left + right) // 2
            if sorted_list[mid] < x:
                left = mid + 1
            else:
                right = mid
        result.append(left)
        sorted_list.insert(left, x)
    return result[::-1]

nums = [5, 2, 6, 1]
print(f"Mảng gốc: {nums}")
print(f"Kết quả:  {count_smaller_binary_search(nums)}")
```

```python
#W7A10
import bisect

def maximumSum(nums, m):
    prefix_sums = []
    max_val = 0
    current_sum = 0

    for x in nums:
        current_sum = (current_sum + x) % m
        max_val = max(max_val, current_sum)
        idx = bisect.bisect_right(prefix_sums, current_sum)

        if idx < len(prefix_sums):
            max_val = max(max_val, (current_sum - prefix_sums[idx] + m) % m)
        bisect.insort(prefix_sums, current_sum)

    return max_val

print(f"Kết quả cho [3, 3, 9, 9, 5] mod 7 là: {maximumSum([3, 3, 9, 9, 5], 7)}")
```

```python
#W7A13
def findLHS(arr):
    counts = {}
    for num in arr:
        counts[num] = counts.get(num, 0) + 1

    max_length = 0

    for x in counts:
        if x + 1 in counts:
            current_length = counts[x] + counts[x + 1]
            max_length = max(max_length, current_length)

    return max_length

input_arr = [1, 3, 2, 2, 5, 2, 3, 7]
print(f"Đầu ra: {findLHS(input_arr)}")
```
