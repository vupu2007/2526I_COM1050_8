```python
#W9A1
def merge_hai_danh_sach(list1, list2):
    res = []
    i = 0
    j = 0

    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            res.append(list1[i])
            i += 1
        else:
            res.append(list2[j])
            j += 1

    res.extend(list1[i:])
    res.extend(list2[j:])

    return res


danh_sach_1 = list(map(int,input("Danh sách 1 : ").split()))
danh_sach_2 = list(map(int,input("Danh sách 2 : ").split()))

ket_qua = merge_hai_danh_sach(danh_sach_1, danh_sach_2)
print("Kết quả trộn:", ket_qua)
```

```python
#W9A2
def count_less_equal(matrix, target, n, m):
    count = 0
    row = n - 1
    col = 0
    while row >= 0 and col < m:
        if matrix[row][col] <= target:
            count += (row + 1)
            col += 1
        else:
            row -= 1
    return count

def find_kth_element(matrix, k, find_largest=True):
    n = len(matrix)
    m = len(matrix[0])

    if find_largest:
        target_k = (n * m) - k + 1
    else:
        target_k = k

    low = matrix[0][0]
    high = matrix[n-1][m-1]
    ans = low

    while low <= high:
        mid = (low + high) // 2
        if count_less_equal(matrix, mid, n, m) >= target_k:
            ans = mid
            high = mid - 1
        else:
            low = mid + 1
    return ans

matrix = [
    [1,  5,  9],
    [10, 11, 13],
    [12, 13, 15]
]
k = 2
print(f"Số lớn thứ {k} là:", find_kth_element(matrix, k))
```

```python
#W9A3
def tim_so_nho_nhat(n):
    s = list(str(n))
    length = len(s)

    for i in range(length - 1):
        idx_min = i
        for j in range(length - 1, i, -1):
            if s[j] < s[idx_min]:
                if i == 0 and s[j] == '0':
                    continue
                if idx_min == i or s[j] < s[idx_min]:
                    idx_min = j

        if idx_min != i:
            s[i], s[idx_min] = s[idx_min], s[i]
            break

    return "".join(s)

so_ban_dau = int(input("Nhập số : "))
print(f"Số nhỏ nhất sau 1 lần hoán đổi: {tim_so_nho_nhat(so_ban_dau)}")
```

```python
#W9A5
n = int(input("Nhap so phan tu list1: "))
list1 = list(map(int, input("Nhap list1: ").split()))

m = int(input("Nhap so phan tu list2: "))
list2 = list(map(int, input("Nhap list2: ").split()))

def is_subarray(list1, list2):
    n, m = len(list1), len(list2)
    if n > m:
        return False
    for i in range(m - n + 1):
        if list2[i:i+n] == list1:
            return True
    return False

if is_subarray(list1, list2):
    print("YES")
else:
    print("NO")

```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0) or (flowerbed[i - 1] == 0)
            right = (i == n - 1) or (flowerbed[i + 1] == 0)
            if left and right:
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True
    return k <= 0
flowerbed = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A8
def is_path_crossing(moves: str) -> bool:
    x, y = 0, 0
    visited = {(0, 0)}

    for move in moves:
        if move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
        elif move == 'U':
            y += 1
        elif move == 'D':
            y -= 1

        if (x, y) in visited:
            return True

        visited.add((x, y))

    return False


moves = input().strip()
print(is_path_crossing(moves))
```

```python
#W9A9
def Median(nums):
    a, b, c, d, e = nums

    # 1–2
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c

    # 3
    if a > c:
        a, c = c, a
        b, d = d, b

    # 4
    if b > e:
        b, e = e, b

    # 5
    if b > c:
        b, c = c, b

    # 6
    if c > e:
        c, e = e, c

    # c là số trung vị
    return c

nums = list(map(int, input().split()))
print(Median(nums))
```

```python
#W9A10
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()

    i = 0
    res = 0

    for h in houses:
        while i + 1 < len(heaters) and abs(heaters[i + 1] - h) <= abs(heaters[i] - h):
            i += 1

        res = max(res, abs(heaters[i] - h))

    return res


houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
